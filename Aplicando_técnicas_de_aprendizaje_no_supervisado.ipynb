Agrupamiento con K-Means

Usamos K-Means para encontrar agrupamientos naturales en el conjunto de datos Iris.

Vamos a agrupar los datos en 3 clusters. El algoritmo intenta minimizar la distancia entre los puntos y sus respectivos centroides. La visualización nos ayuda a comprender cómo se separaron los datos en grupos distintos.

```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Cargar Dataset
iris = load_iris()
X = iris.data

# Aplicar K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

# Visualizar los clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.title("Agrupamiento con K-Means")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

Agrupamiento Jerárquico

Ahora, vamos a aplicar un Agrupamiento Jerárquico y visualizar las relaciones entre los grupos con un dendrograma.

```python
from sklearn.datasets import load_iris
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data

# Normalizar los datos para mejorar el rendimiento del clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear el dendrograma
plt.figure(figsize=(10, 5))
sch.dendrogram(sch.linkage(X_scaled, method='ward'))
plt.title("Dendrograma del Agrupamiento Jerárquico")
plt.xlabel("Muestras")
plt.ylabel("Distancia Euclidiana")
plt.show()
```

El dendrograma muestra cómo se forman los clusters, permitiendo una elección más intuitiva del número ideal de agrupamientos.

Cada línea en el eje X representa una muestra del conjunto de datos. En el caso del conjunto de datos Iris, cada muestra es una flor. A su vez, el eje Y indica la distancia euclidiana (u otra métrica) entre los grupos que se están uniendo en el proceso de clustering.

Podemos observar cómo los datos se organizan en diferentes niveles. Dependiendo del corte que hacemos en el árbol, podemos identificar un número mayor o menor de grupos.

El eje X puede volverse un poco confuso en la representación, porque el dendrograma intenta representar todas las muestras. Cuando tenemos muchos datos, las etiquetas pueden superponerse. Sin embargo, esta superposición no afecta el análisis, ya que lo más importante es observar la estructura de las conexiones entre los grupos, y no las etiquetas individuales.

Al principio, cada muestra es un cluster separado. A medida que subimos en el gráfico, los clusters comienzan a unirse. Cuanto mayor sea la altura de la fusión entre dos grupos, más diferentes son. El gran salto en medio del gráfico indica la fusión de grandes grupos distintos.

¿Cómo podemos definir el número de clusters a partir de esta representación?

Podemos imaginar un corte horizontal en el gráfico. El número de líneas verticales cortadas en ese punto representa los clusters finales.
Si cortamos a la altura ≈10 en el eje Y, obtendríamos tres grupos principales (verde, naranja y azul).
Ahora, vamos a entender cómo extraer los clusters de forma automática con el algoritmo HCA (Agrupamiento Jerárquico Aglomerativo). Es un método de agrupamiento que construye una jerarquía de clusters. Comienza considerando cada punto de datos como un cluster separado y, paso a paso, va uniendo los más cercanos hasta que todos los puntos estén agrupados en un único cluster.

En este caso, ya sabemos, a partir del análisis del dendograma, cómo se pueden agrupar los datos, así que ya tenemos un número adecuado de clusters para la aplicación del algoritmo de agrupamiento.

```python
from sklearn.cluster import AgglomerativeClustering

# Aplicar Agrupamiento Jerárquico definiendo 3 grupos
hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
clusters = hc.fit_predict(X_scaled)

# Visualizar los clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.title("Agrupamiento con Agrupamiento Jerárquico")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

Reducción de Dimensionalidad con PCA

Por último, aplicamos PCA para reducir las 4 dimensiones del conjunto de datos Iris a solo 2, facilitando la visualización de los datos.

El PCA nos ayuda a visualizar la separación entre las especies de flores, incluso en un espacio de dos dimensiones, manteniendo la mayor parte de la varianza de los datos originales.

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target  # Etiquetas de las especies de flores

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualizar los datos reducidos
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("Reducción de Dimensionalidad con PCA")
plt.colorbar(label="Especies de Flores")
plt.show()
```
