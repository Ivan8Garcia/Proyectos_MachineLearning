Después de esta clase, ya entiendes el concepto de redes neuronales y cómo usar modelos preentrenados para tareas de clasificación de imágenes. ¡Ahora es el momento de poner todo esto en práctica!

Vas a usar el modelo MobileNetV2, que fue entrenado en el conjunto de datos ImageNet, para clasificar imágenes de objetos que envíes.

Lo que debes hacer en esta actividad:

Usar el modelo MobileNetV2 para predecir la clase de un objeto en una imagen.
Preprocesar la imagen para el formato correcto que el modelo espera.
Mostrar la clase identificada en la imagen.

Con MobileNetV2, podemos cargar una imagen y rápidamente identificar su contenido. El preprocesamiento adecuado de la imagen, que incluye redimensionamiento y normalización, garantiza que el modelo reciba los datos en la forma correcta para proporcionar una predicción precisa.

Paso 1: Cargar y preparar la imagen

Primero, harás la carga de la imagen y la prepararás para el modelo. El modelo espera que la imagen tenga un tamaño de 224x224 píxeles, así que redimensionaremos la imagen a ese tamaño.

Paso 2: Cargar el modelo preentrenado

El MobileNetV2 ya ha sido entrenado en una gran base de datos de imágenes, lo que lo hace capaz de clasificar una gran variedad de objetos. Vamos a cargar este modelo y usarlo para hacer predicciones.

Paso 3: Preprocesamiento de la imagen

El preprocesamiento garantiza que la imagen esté en el formato esperado por el modelo. Necesitamos normalizar los valores de los píxeles y hacer otras transformaciones para que el modelo pueda hacer la predicción de manera efectiva.

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
import numpy as np
import cv2
from google.colab import files
from PIL import Image

# Hacer upload de la imagen
uploaded = files.upload()
imagen_path = list(uploaded.keys())[0]  # Obtiene el nombre del archivo enviado

# Cargar la imagen usando OpenCV
imagen = cv2.imread(imagen_path)
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Cargar el modelo MobileNetV2 preentrenado
modelo = MobileNetV2(weights="imagenet")

# Preprocesar la imagen para el formato esperado por el modelo
imagen_redimensionada = cv2.resize(imagen_rgb, (224, 224))
imagen_array = np.expand_dims(imagen_redimensionada, axis=0)
imagen_array = preprocess_input(imagen_array)

# Hacer la predicción
predicciones = modelo.predict(imagen_array)
label = decode_predictions(predicciones)
print("Objeto identificado:", label[0][0][1])  # Muestra la clase identificada
```
