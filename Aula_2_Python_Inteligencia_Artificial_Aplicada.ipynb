##crear una funcion que va a recibir como argumento una lista de emails y le va a pedir a una IA que los resuma en una sola linea

```python
cuerpos_emails = [
    "Hola, te envío el reporte de ventas del trimestre que solicitaste en la reunión. Quedo atento a tus comentarios.",
    "¡Tu suscripción ha sido renovada con éxito! El cargo de 9.99€ se ha aplicado a tu tarjeta terminada en 4421.",
    "¿Qué tal todo? Hace mucho que no hablamos. Podríamos quedar a tomar un café este jueves si tienes un hueco.",
    "Estimado alumno, le recordamos que el plazo para la entrega del proyecto final finaliza este viernes a medianoche.",

]
```

```python
def resumir_emails(cuerpos_emails):
  for numero, email in enumerate(cuerpos_emails):
    respuesta = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=f"""Te voy a enviar el cuerpo de un email y necesito que lo resumas en una sola linea. Dejando clara la intencion de dicho correo.
         El correo es el siguiente:{email}""")

    print(f"E-mail N° {numero + 1}: {respuesta.text}")
```

```python
resumir_emails(cuerpos_emails)
```

```python
def resumir_emails(cuerpos_emails):
    lista_resumenes = []
    for numero, email in enumerate(cuerpos_emails):
        respuesta = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {
                    "role": "user",
                    "content": f"""Te voy a enviar el cuerpo de un correo electrónico. Quiero que lo resumas en solo 1 línea,
indicando la intención de dicho correo. El correo es el siguiente: {email}"""
                }
            ]
        )
        texto= respuesta.choices[0].message.content

        print(f"E-Mail N° {numero + 1}: {respuesta.texto}")
        lista_resumenes.append(f"E-Mail N° {numero + 1}: {respuesta.texto}")

    return lista_resumenes

```

```python
lista_resumen=resumir_emails(cuerpos_emails)
```
