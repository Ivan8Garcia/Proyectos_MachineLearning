```python
#Etapa 1
lista_de_preguntas = [
    "¿Cuál es la capital de Francia?",
    "¿En qué año se fundó Google?",
    "¿Cuál es el río más largo del mundo",
    "¿Cuál es el planeta más grande del sistema solar?",
    "¿Quién pintó la Mona Lisa?"
]
```

```python
#creacion de archivo txt, PASO #1. LISTO

with open("lista-de-preguntas.txt", "w", encoding="utf-8") as archivo:
  for pregunta in lista_de_preguntas:
    archivo.write(pregunta+ "\n")
```

```python
# PASO 2

lista_desafio=[]
with open("lista-de-preguntas.txt", "r", encoding="utf-8") as archivo:
  for linea in archivo:
    lista_desafio.append(linea.strip())
```

```python
print(lista_desafio)
```

```python
#PASO 3
#Etapa 3
lista_de_diccionarios_respuestas = []

for pregunta in lista_desafio:
    respuesta = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
              "role": "user",
              "content": f"""te voy a enviar una pregunta y necesito que me duelvas la respuesta corta. pregunta:{pregunta}"""
             }
      ]

  )
  texto=respuesta.choices[0].message.content
  lista_de_diccionarios_respuestas.append({"pregunta":pregunta,"respuesta":texto})
```

```python
#PASO 4

  with open("lista-de-respuestas.csv", "w", encoding="utf-8") as archivo:
    archivo.write("pregunta,respuesta\n")
    for diccionario in lista_de_diccionarios_respuestas:
      archivo.write(f"{diccionario['pregunta']},{diccionario['respuesta']}\n")
```

```python
import pandas as pd
import random

# Definir posibles valores para las columnas
nombres_productos_posibles = [
    'Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Auriculares', 'Teclado', 'Mouse', 'Monitor', 'Impresora', "Altavoz",
    'Cámara', 'Disco Duro', 'Router', 'Webcam', 'Micrófono', 'Gafas VR', 'Consola', 'Control Remoto',"Proyector", "Bateria externa"
]
categorias_posibles = ['Electrónica', 'Hogar', 'Oficina', 'Entretenimiento', 'Accesorios']

# Seleccionar 10 nombres de productos únicos aleatoriamente
productos_elegidos = random.sample(nombres_productos_posibles, 10)

# Generar datos para las 10 filas
data = {
    'Nombre de producto': productos_elegidos,
    'Categoria': [random.choice(categorias_posibles) for _ in range(10)],
    'Precio': [round(random.uniform(50, 1500), 2) for _ in range(10)],
    'Stock': [random.randint(10, 100) for _ in range(10)]
}

# Crear el DataFrame
df_productos = pd.DataFrame(data)

# Mostrar el DataFrame
display(df_productos)
```
